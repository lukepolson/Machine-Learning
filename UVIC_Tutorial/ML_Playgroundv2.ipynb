Now extract the features and labels from the data frame into np.arrays.

### IMPORTANT

The labels (STAR = 6, GALAXY = 3) are not standard label values. Normally, binary classification expects the classes to be 0 or 1. So we need to rescale those labels, as done below.

```python
# FROM THE DATA FRAME EXTRACT THE LABELS AND FEATURES
# CHECK THE SHAPE (SIZE) OF EACH
labels = dfSG.type_r.values
labels = (labels - 3)/3
labels = labels.reshape(labels.shape[0],1)

features = dfSG.drop(["type_r"], axis = 1).values
print('Shape of labels: ',labels.shape)
print('Shape of features: ',features.shape)
```

### Rescaling the features

The numerical values of the features differ significantly. Using them directly in the network would result in the network learning only from a subset of these features (those with dominant values) and disregarding the others. Therefore, it is important to rescale the features so they have similar numerical values without losing their information content.

In the code cell below, there are two independent methods presented: (i) min-max rescaling, and (ii) rescaling so the mean is 0 and variance is 1 (unit normalize)

```python
# MIN-MAX SCALING
# rescale each feature between its min and max values
max_vect = np.amax(features,axis=0,keepdims=True)
min_vect = np.amin(features,axis=0,keepdims=True)
norm_features = (features - min_vect)/(max_vect - min_vect)

# UNIT NORMAL DISTRIBUTIONS
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# norm_features = sc.fit_transform(features)

# Check the resclaing
plt.hist(norm_features[:,2],bins=20)
```

## Train - Validate - Test sets

Next split the data into training, validation and test sets. We will follow the usual prescription of 80:10:10 ratio for this split.

####**NOTE** Scikit-learn does only a two-way split, so do this in two steps

```python
from sklearn.model_selection import train_test_split

# IN TWO STEPS, SPLIT DATA INTO TRAIN, VALIDATION AND TEST SETS
feat_train, feat_valtest, lab_train, lab_valtest = train_test_split(norm_features, labels, test_size = 0.2, random_state = 42)
feat_val, feat_test, lab_val, lab_test = train_test_split(feat_valtest, lab_valtest, test_size = 0.5, random_state = 42)

print('Shape of train set: ',feat_train.shape)
print('Shape of validate set: ',feat_val.shape)
print('Shape of test set: ',feat_test.shape)
print('Shape of train labels: ',lab_train.shape)

train_siz = feat_train.shape[0]
val_siz = feat_val.shape[0]
test_siz = feat_test.shape[0]
n_feat = feat_train.shape[1]
print('Sizes of train, validate and test sets: %d, %d, %d' % (train_siz,val_siz,test_siz))
print('Number of features: %d' % (n_feat))
```

# Build a fully connected network using PyTorch

```python
# IMPORT PYTORCH AND THE NECESSARY ASSOCIATED PACKAGES
import torch
from torch.autograd import Variable
```

Let's first define the architecture, number of layers, number of neurons in each layer, etc

*   Nsamp = number of samples (in train set or batch)
*   Dim_in = size of first hidden layer (number of features + 1)
*   Dim_out = size of output (=1 for our binary classification case)

```python
# Choose batch size 1/100 of train set, dim_in = num of features
batch_siz = feat_train.shape[0]/100
n_feat = feat_train.shape[1]
hiddn1 = 8
hiddn2 = 4
Nsamp, Dim_in, Dim_out = batch_siz,n_feat,1
print('Chosen network params: Nsamp, Dim_in, Dim_out ',Nsamp, Dim_in, Dim_out)
```

```python
# Using PyTorch torch.nn package is a good first start
# For more complex models, we can define classes
# The modules are put together like Lego blocks


# Basic: Linear + ReLU/sigmoid
PyTmodel = torch.nn.Sequential(
      torch.nn.Linear(Dim_in, hiddn1),
      torch.nn.ReLU(),
      torch.nn.Dropout(0.3),
      torch.nn.Linear(hiddn1, hiddn2),
      torch.nn.ReLU(),
      torch.nn.Linear(hiddn2, Dim_out),
      torch.nn.Sigmoid(),
)
```

```python
# Various loss functions are available in torch.nn
# loss_fn = torch.nn.MSELoss(reduction='sum')
# loss_fn = torch.nn.CrossEntropyLoss()
loss_fn = torch.nn.BCELoss()
```

```python
# torch.optim also provides various optimizers
learning_rate = 0.2
# optimizer = torch.optim.SGD(PyTmodel.parameters(), lr=learning_rate, momentum=0.9)
optimizer = torch.optim.Adam(PyTmodel.parameters(), lr=learning_rate)
```

```python
# We first need to convert numpy arrays to Torch tensors
feat_train_PyT = torch.from_numpy(feat_train) # (N samples, D_in features)
lab_train_PyT = torch.from_numpy(lab_train)

# Run the training loop
niter = 5000
nchks = 500
for itr in range(niter):
    # Forward pass
    y_pred = PyTmodel(Variable(feat_train_PyT.float()))
    # Loss
    loss = loss_fn(y_pred, Variable(lab_train_PyT.float()))
    if ((itr % nchks) == 0):
        print(itr, loss.item())
    # Run the back prop (using optimizer)
    optimizer.zero_grad()
    loss.backward()
    # optimizer.step()
    with torch.no_grad():
      for param in PyTmodel.parameters():
        param -= learning_rate * param.grad
```

## TEST THE TRAINED NETWORK

Run the network on the test dataset. Print out the predictions and compare with the test labels. Then compute the _Confusion Matrix_ and various performance metrics. Also plot the confusion matrix.

###NOTE

Since the Sigmoid function outputs only a probability, we can choose the decision boundary, _db_ between the two classes.

```python
# Create predictions for the test set
# set the decision boundary
db = 0.6
feat_test_PyT = torch.from_numpy(feat_test)
y_pred_PyT = PyTmodel(Variable(feat_test_PyT.float()))
y_pred = (y_pred_PyT > db).float().numpy()

# Print out the first ten predictions 
# and compare with the test labels
for i in range(10):
  print(lab_test[i],y_pred[i])
```

```python
# Compute the confusion matrix for any specified decision boundary
def comp_cm(yprob,lab_test,dbry):
    # initialize the indices
    TP,FP,FN,TN = 0,0,0,0

    # check if input sizes match
    if (yprob.shape[0] != lab_test.shape[0]):
        print('Predicted and Test y shapes do not match: %d %d' % (yprob.shape[0],lab_test.shape[0]))
        return TP,FP,FN,TN
    
    # All ok, continue
    ypred = np.zeros((yprob.shape[0],1))
    tru_indx = np.where(yprob > dbry)[0]
    ypred[tru_indx] = 1
    for indx in range(lab_test.shape[0]):
        if ((lab_test[indx] == 1) and (ypred[indx] == 1)):
            TP = TP+1
        elif ((lab_test[indx] == 0) and (ypred[indx] == 1)):
            FP = FP+1
        elif ((lab_test[indx] == 1) and (ypred[indx] == 0)):
            FN = FN+1
        else: # ((lab_test[indx] == 0) and (ypred[indx] == 0))
            TN = TN+1
    # cmat = np.array([[TP,FN],[FP,TN]])
    cmat = np.array([[TN,FP],[FN,TP]])
    # cmat_norm = cmat/(float(TP+FP+FN+TN))
    cmat_norm = cmat.astype('float') / cmat.sum(axis=1)[:, np.newaxis]
    print('Total cmat: %d' % np.sum(cmat))
    print('Total cmat norm: %d' % np.sum(cmat_norm))
    
    # return indices and cm
    return TP,FP,FN,TN,cmat,cmat_norm
```

```python
# SET THE DECISION BOUNDARY HERE
# CHOOSE ANY VALUE BETWEEN 0 AND 1, AND SEE HOW THE PERFORMANCE CHANGES
dbry = 0.6

TP,FP,FN,TN,cmat,cmat_norm = comp_cm(y_pred,lab_test,dbry)
print('TP,FN,FP,TN: ',TP,FN,FP,TN)
print('Accuracy: %.2f' % ((TP+TN)/lab_test.shape[0]))
```

```python
def print_confusion_matrix(confusion_matrix, class_names, figsize = (10,7), fontsize=14):
    """Prints a confusion matrix, as returned by sklearn.metrics.confusion_matrix, as a heatmap.
    
    Arguments
    ---------
    confusion_matrix: numpy.ndarray
        The numpy.ndarray object returned from a call to sklearn.metrics.confusion_matrix. 
        Similarly constructed ndarrays can also be used.
    class_names: list
        An ordered list of class names, in the order they index the given confusion matrix.
    figsize: tuple
        A 2-long tuple, the first value determining the horizontal size of the ouputted figure,
        the second determining the vertical size. Defaults to (10,7).
    fontsize: int
        Font size for axes labels. Defaults to 14.
        
    Returns
    -------
    matplotlib.figure.Figure
        The resulting confusion matrix figure
    """
    df_cm = pd.DataFrame(
        confusion_matrix, index=class_names, columns=class_names, 
    )
    colormap = plt.cm.bwr
    fig = plt.figure(figsize=figsize)
    try:
        heatmap = sns.heatmap(df_cm, annot=True, fmt="d",cmap=plt.cm.Blues)
    except ValueError:
        raise ValueError("Confusion matrix values must be integers.")
    heatmap.yaxis.set_ticklabels(heatmap.yaxis.get_ticklabels(), rotation=0, ha='right', fontsize=fontsize)
    heatmap.xaxis.set_ticklabels(heatmap.xaxis.get_ticklabels(), rotation=45, ha='right', fontsize=fontsize)
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    return fig
```
