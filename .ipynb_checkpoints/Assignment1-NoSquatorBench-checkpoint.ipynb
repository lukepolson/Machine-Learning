# Part 1

## Read in the Data

```python
df = pd.read_csv('https://github.com/sstangl/openpowerlifting-static/raw/gh-pages/openpowerlifting.csv.bz2')
df = df[df['Equipment']=='Raw']
df = df[['Sex', 'Age', 'Tested', 'BodyweightKg', 'Best3SquatKg', 'Best3BenchKg', 'Best3DeadliftKg']]
df = df.dropna(subset=['Sex', 'Age', 'Tested', 'Best3SquatKg', 'Best3BenchKg', 'Best3DeadliftKg'])
df = df[(df['Best3SquatKg']>0) & (df['Best3BenchKg']>0) & (df['Best3DeadliftKg']>0)]
df.sort_index(inplace=True)
```

## Stratified Sampling

Taking into account the histogram above, we probably want catagory 1 and 100kg, catagory 4 at 300kg, and anything above 300kg at catagory 5. We can use a linear mapping to achieve this. We have

$$1 = 100m+b $$
$$4 = 300m+b $$

and thus $b=-1/2$ and $m=3/200$. Applying this mapping to our data set yields.

```python
df['deadlift_cat'] = np.ceil((3/200)*df['Best3DeadliftKg']-1/2)
df['deadlift_cat'].where(df['deadlift_cat'] < 5, 5.0, inplace=True)
```

Plot to show how many entries are in each catagory.

```python
df['deadlift_cat'].hist(bins=20)
plt.show()
```

Now we do stratified sampling on the data set to ensure we get equal proportions from each catagory.

```python
split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=42)
for train_index, test_index in split.split(df, df['deadlift_cat']):
    strat_train_set = df.iloc[train_index]
    strat_test_set= df.iloc[test_index]
```

# Part 3

We seperate the predictors from the labels:

```python
lifting = strat_train_set.drop(['Best3DeadliftKg', 'deadlift_cat'], axis=1)
lifting_labels = strat_train_set['Best3DeadliftKg'].copy()
```

Now we seperate the numerical data from the categorical data.

```python
lifting_num = lifting.drop(['Sex', 'Tested'], axis=1)
lifting_cat = lifting[['Sex', 'Tested']]
```

Get the catagory headers.

```python
num_attribs = list(lifting_num)
cat_attribs = list(lifting_cat)
```

## Creating Pipelines

Create classes that take data from a DataFrame and turn it into a NumPy array, and add additional attributes.

```python
class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values
    
bench_ix = 3; squat_ix = 2
class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bench_per_squat = True):
        self.add_bench_per_squat = add_bench_per_squat
    def fit(self, X, y=None):
        return self
    def transform(self, X, y=None):
        bench_per_squat = X[:, bench_ix] / X[:, squat_ix] # this might be used
        if self.add_bench_per_squat:
            return np.c_[X, bench_per_squat]
        else:
            return np.c_[X]
```

Design the pipeline

```python
num_pipeline = Pipeline([
    ('selector', DataFrameSelector(num_attribs)),
    ('imputer', impute.SimpleImputer(strategy="median")),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler())
])

cat_pipeline = Pipeline([
    ('selector', DataFrameSelector(cat_attribs)),
    ('cat_encoder', OneHotEncoder())
])

full_pipeline = FeatureUnion(transformer_list =[
    ("num_pipeline", num_pipeline),
    ("cat_pipeline", cat_pipeline),
])
```

Pass the data through the pipeline.

```python
lifting_prepared = full_pipeline.fit_transform(lifting)
```

# Part 4

A little function to display the scores:

```python
def display_scores(scores):
    scores = np.sqrt(-scores)
    print('Scores:', scores)
    print('Mean:', scores.mean())
    print('STD:', scores.std())
```

## The Three Models

### Linear Regression

Train the model.

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(lifting_prepared, lifting_labels)
```

See how it performs on some test data.

```python
some_data = lifting.iloc[:5]
some_labels = lifting_labels.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)

print('Predictions:', lin_reg.predict(some_data_prepared))
print('Actual Values:', some_labels.values)
```

Perform some cross-validation and see how well it works.

```python
scores = cross_val_score(lin_reg, lifting_prepared, lifting_labels,
                        scoring='neg_mean_squared_error', cv=3)
display_scores(scores)
```

### Decision Tree Regressor

Train the model.

```python
from sklearn.tree import DecisionTreeRegressor
tree_reg = DecisionTreeRegressor()
tree_reg.fit(lifting_prepared, lifting_labels)
```

See how it performs on some test data.

```python
some_data = lifting.iloc[:5]
some_labels = lifting_labels.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)

print('Predictions:', tree_reg.predict(some_data_prepared))
print('Actual Values:', some_labels.values)
```

Perform some cross-validation and see how well it works.

```python
scores = cross_val_score(tree_reg, lifting_prepared, lifting_labels,
                        scoring='neg_mean_squared_error', cv=3)
display_scores(scores)
```

### Random Forest Regressor

```python
from sklearn.ensemble import RandomForestRegressor
forest_reg = RandomForestRegressor(n_estimators=10)
forest_reg.fit(lifting_prepared, lifting_labels)
```

See how well it performs on some test data.

```python
some_data = lifting.iloc[:5]
some_labels = lifting_labels.iloc[:5]
some_data_prepared = full_pipeline.transform(some_data)

print('Predictions:', forest_reg.predict(some_data_prepared))
print('Actual Values:', some_labels.values)
```

Perform some cross-validation and see how well it works.

```python
scores = cross_val_score(forest_reg, lifting_prepared, lifting_labels,
                        scoring='neg_mean_squared_error', cv=3)
display_scores(scores)
```
