The values are as follows:

* **cp: chest pain type**
    * Value 0: asymptomatic
    * Value 1: atypical angina
    * Value 2: non-anginal pain
    * Value 3: typical angina
    
* **trestbps: resting blood pressure**
    * Measured in mmHg
    
* **chol: serum cholestoral**
    * measured in in mg/dl

* **fbs: fasting blood sugar**
    * 1 if greater than >120mg/dl; else 0
    
* **restecg: resting electrocardiographic results**
     * Value 0: showing probable or definite left ventricular hypertrophy by Estes' criteria
     * Value 1: normal
     * Value 2: having ST-T wave abnormality (T wave inversions and/or ST elevation or depression of > 0.05 mV)
     
* **thalach: max heart rate achieved**
    * Measured in beats per minute
    
* **exang: exercise induced angina**
    * 1 if true else 0
     
* **slope: the slope of the peak exercise ST segment**
    * 0: downsloping 
    * 1: flat 
    * 2: upsloping
    
* **oldpeak: ST depression induced by exercise relative to rest**
    * This has to do with heart pulses on heart monitor being below expected value
    
* **ca: Number of major blood vessels covereed by fluoroscopy**
    * 
    
* **thal: Radioactive tracer test**
    * 1 = fixed defect
    * 2 = normal
    * 7 = reversable defect

* **target (maybe THE most important feature)** 
    * 0 = disease
    * 1 = no disease

# Prepare the Data


Get train and test

```python
def split_train_test(data, test_ratio):
    shuffled_indices = np.random.permutation(len(data))
    test_set_size = int(len(data)*test_ratio)
    
    test_indices = shuffled_indices[: test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    
    return data.iloc[train_indices], data.iloc[test_indices]
```

WILL PROBABLY WANT TO SPLIT BY AGE RANGE AS WELL

```python
from sklearn.model_selection import StratifiedShuffleSplit

split = StratifiedShuffleSplit(n_splits=1, test_size=0.2, random_state=40)
for train_index, test_index in split.split(df, df['sex']): #this is only a single loop
    df_train= df.loc[train_index] #train_index is an array
    df_test= df.loc[test_index] # test index is an array
```

## Get labels

```python
y_train = np.array(df_train['target'])
y_test = np.array(df_test['target'])
```

## Get predictors

```python
attribs = df.columns.values[0:-1]
```

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.base import BaseEstimator, TransformerMixin
from sklearn import impute
```

Standard Scale the data, add attributes, etc..


```python
class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values

pipeline = Pipeline([
    ('selector', DataFrameSelector(attribs)),
    ('imputer', impute.SimpleImputer(strategy="median")),
    # ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler())
])

X_train = pipeline.fit_transform(df_train)
X_test = pipeline.fit_transform(df_test)
```

# Train Some Models

Trained purely on raw data.

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import VotingClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# Train 3 different classifiers
log_clf = LogisticRegression(solver="liblinear", random_state=42)
rnd_clf = RandomForestClassifier(n_estimators=10, random_state=42)
svm_clf = SVC(gamma="auto", random_state=42, probability = True)
 
# HERE IS CODE FOR VOTING CLASSIFIER
voting_clf = VotingClassifier(
    estimators=[('lr', log_clf), ('rf', rnd_clf), ('svc', svm_clf)],
    voting='soft')
```

Print outcomes.

```python
from sklearn.metrics import accuracy_score

for clf in (log_clf, rnd_clf, svm_clf, voting_clf):
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    print(clf.__class__.__name__, accuracy_score(y_test, y_pred))
```

# Tune Each Model

## SVC

```python
from sklearn.svm import SVC
from sklearn.model_selection import RandomizedSearchCV
from scipy.stats import reciprocal, uniform
```

Search over parameters.

```python
param_distributions = {"gamma": reciprocal(0.005, 0.05), "C": uniform(0, 40)}
SVC_search_cv = RandomizedSearchCV(SVC(), param_distributions, n_iter=30, verbose=2, cv=10, random_state=42)
SVC_search_cv.fit(X_train, y_train)
```

Get results.

```python
results = SVC_search_cv.cv_results_

Cs = np.array(results['param_C'], dtype=float)
gammas = np.array(results['param_gamma'], dtype=float)
scores = np.array(results['mean_test_score'], dtype=float)
```

Plot which are best

```python
plt.scatter(Cs, gammas, c=scores, cmap="jet", lw=0)
plt.xlabel('C')
plt.ylabel(r'$\epsilon$')
plt.colorbar(label='Score')
plt.title('Scores')
plt.show()
```
