```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os

from sklearn.metrics import precision_score

import tensorflow as tf
from tensorflow import keras
```

# Load Data and Get Train and Test

```python
data = keras.datasets.mnist.load_data()
```

Get train_full and test

```python
(X_train_full, y_train_full), (X_test, y_test) = data
X_train_full = X_train_full / 255.0; X_test = X_test / 255.0
```

Split train_full into train and validation

```python
X_valid, X_train = X_train_full[:5000], X_train_full[5000:]
y_valid, y_train = y_train_full[:5000], y_train_full[5000:]
```

# Generate Model

## Build Model

```python
model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(300, activation="relu"),
    keras.layers.Dense(100, activation="relu"),
    keras.layers.Dense(10, activation="softmax") ])
```

## Compile Model

```python
model.compile(loss='sparse_categorical_crossentropy', optimizer=keras.optimizers.SGD(lr=1e-3))
```

## Train Model

### Callbacks

Also save checkpoints. First initialize tensorboard stuff.

```python
root_logdir=os.path.join(os.curdir, "my_logs")

def get_run_logdir():
    import time
    run_id = time.strftime("run_%Y_%m_%d-%H_%M_%S")
    return os.path.join(root_logdir, run_id)

run_logdir = get_run_logdir()
```

Initialize callbacks.

```python
# Callbacks
checkpoint_cb = keras.callbacks.ModelCheckpoint("my_keras_MNIST_model.h5", save_best_only=True)
early_stopping_cb = keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)
tensorboard_cb = keras.callbacks.TensorBoard(run_logdir)
```

Fit model.

```python
history = model.fit(X_train, y_train, epochs=1000, validation_data=(X_valid, y_valid),
                   callbacks=[checkpoint_cb, early_stopping_cb, tensorboard_cb])
```

## Check Precision

```python
y_pred = model.predict_classes(X_test)
```

```python
precision_score(y_test, y_pred, average='macro')
```

# Learning Rate Stuff

```python
test_model = keras.models.Sequential([
    keras.layers.Flatten(input_shape=[28, 28]),
    keras.layers.Dense(300, activation="relu"),
    keras.layers.Dense(100, activation="relu"),
    keras.layers.Dense(10, activation="softmax") ])
```

Compile with a "decay" rate that grows each time.

```python
sgd = keras.optimizers.SGD(lr=1e-3, decay = np.exp(np.log(10**6)/500))
test_model.compile(loss='sparse_categorical_crossentropy', optimizer=sgd)
```

Fit model. No callbacks needed here.

```python
test_history = test_model.fit(X_train, y_train, epochs=500, validation_data=(X_valid, y_valid))
```
